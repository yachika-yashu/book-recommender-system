#Popularity based recommender system:
that is based on whole platforms ratings of books, or you can say top 10 recommended books BASED on highest average rating also we will make sure at least 250 people have given the voting

merge books and ratings csv on the basis of isbn, need book name and ratings

```python
merged_list=books.merge(ratings, on='ISBN').groupby('Book-Title')['Book-Rating'].count().reset_index()  #df groupby(col1)[col2].agg() col1 is the col to which we have to group, col 2 is where to apply calculation
merged_list.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
merged_list
```

```python
avg_list=books.merge(ratings, on='ISBN').groupby('Book-Title')['Book-Rating'].mean().reset_index()  #df groupby(col1)[col2].agg() col1 is the col to which we have to group, col 2 is where to apply calculation
avg_list.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_list
```

```python
final_list=merged_list.merge(avg_list, on='Book-Title')
final_list=final_list[final_list['num_ratings']>=250].sort_values(by='avg_ratings',ascending=False).head(50)
final_list
```

now in above list add author, publisher and image

```python
final_list.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Publisher', 'Image-URL-S'	,'num_ratings','avg_ratings']]
```
